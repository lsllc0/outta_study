## 경쟁력 지표 크롤링

**크롤링 방향**



---


크롤링은 하는 방법은 다양하다.가장 먼저 단순하게는, selenium을 이용해서 각 대학의 정보창을 키고, 정보창에서 개별적으로 크롤링을 수행하는 것이다.

하지만, 해당 사이트는 html 코드에 정보를 저장하고 있지 않으며, 정보창을 키는 버튼 또한 html의 버튼 형식으로 구현되어 있지 않다.

다행히, 해당 사이트는 대학별 코드로 각 대학 정보 사이트에 접속할 수 있도록 URL이 구성이 되어있었다.

아래는 가야대학교 제1캠퍼스의 우리대학경쟁력 URL이다.

https://academyinfo.go.kr/popup/pubinfo1690/list.do?schlId=0002748

https://academyinfo.go.kr/popup/pubinfo1690/list.do?schlId=0002749

URL을 보면 우리는 어느정도 URL에 규칙성이 존재한다는 것을 파악할 수 있다. 따라서 우리는 각 URL에 접근만 할 수 있다면, 각 URL에 있는 대학별 경쟁력 지표를 크롤링할 수 있을 것이다.

### 크롤링 가능한 대학코드 가져오기

앞에서 말했듯, selenium이 막힌 상황에서 우리는 URL을 통해서 우회하여 데이터를 수집할 것이다. 하지만, 가야대학교의 경우 제1캠퍼스와 제2캠퍼스가 URL 번호가 상당히 비슷했으나, 그렇다고 모든 대학이 일렬로 쭉 번호가 매겨진 것은 아닌 것 같다.

아래 링크는 URL 기준 6번 대학을 호출하는 URL이다.

https://academyinfo.go.kr/popup/pubinfo1690/list.do?schlId=0000006

들어가서 확인할 수 있듯, 6번 대학에 대한 페이지는 존재하지 않는다.

하지만, 5번 대학은 존재하는 것을 확인할 수 있다.

https://academyinfo.go.kr/popup/pubinfo1690/list.do?schlId=0000005

메인 페이지에서 대학들의 URL 번호를 파악하면, 어느 정도 규칙성이 존재하긴 하지만, 그렇다고 모든 번호에 대해서 정보가 있는 것은 아닌 것 같다. 그렇다면, 우리는 정보가 존재하는 대학의 URL 번호를 미리 파악한다면, 크롤링을 조금 더 빠르고 효율적으로 수행할 수 있을 것이다.

간단히 주어진 URL에 대해서 대학명을 크롤링해보자. 만약 크롤링이 된다면, 해당 URL 번호가 존재하는 것이고, 되지 않으면 해당 URL 번호가 존재하지 않을 것이다.존재하는 것만 번호를 저장해두자.

```python
import requests
from bs4 import BeautifulSoup
import re

#url에 formatting을 할 수 있는 형식으로 만들어줍니다.
url = "https://www.academyinfo.go.kr/popup/pubinfo1690/list.do?schlId={}"

#크롤링 가능한 대학명과 URL 번호를 받을 수 있는 리스트를 만들어줍니다.
univ_list = []


for i in range(1, 333):  # 범위를 원하는 만큼 설정하세요. 대학 URL의 특성을 파악하고 적당한 숫자를 입력하세요.
    #대학명 추출이 성공했을 경우
    try:
        schl_id = str(i).zfill(7) #주어진 숫자를 000xxxx의 7자리 형태로 바꿔줍니다.
        response = requests.get(url.format(schl_id)) #request get 통해 url 가져오기

        if response.status_code != 200: #에러 난 경우
            continue #그대로 진행

        soup = BeautifulSoup(response.text, "html.parser") #html parser로 parsing

        #대학명 추출

        ### 크롤링 코드를 작성해주세요. ###

        schoolname1 = soup.select_one('#contentsWrap > div > div > div.college-info-box > div.college-info-inner > div.info-box > div.college-name-wrap > span')
        schoolname2 = schoolname1.get_text().split()
        college_name = schoolname2[0]


        ### 크롤링한 대학명은 college_name 변수에 저장해주세요. ###

        if college_name: #college_name이 있다면 schl_id와 함께 append
            univ_list.append((college_name,schl_id))
            print(college_name, schl_id) #진척도를 나타내기 위해 print문을 걸어두는 것을 추천합니다.

    #대학명 추출에 실패했을 경우
    except:
        pass

#돌리다가 충분히 대학 코드를 다 가져왔다고 생각하는 시점에 중지해서 크롤링을 중단해주시면 좋습니다.
```

**전처리 방향**


---


데이터를 가져오면, 전처리를 해야하는 부분을 확인할 수 있다.

1) 장학금, 교육비 등 재무 정보
- "원", "," 등을 처리하여 숫자형으로 저장한다.

2) 충원율, 취업률 등 비율 정보
- "%"를 처리하여 숫자형으로 저장한다.

3) 재학생수, 졸업자수 등 명수 정보
- "명", "," 등을 처리하여 숫자형으로 저장한다.

```python
### 아래에 전처리 코드를 작성해주세요. ###

##참고할 만한 함수: df.apply, int, replace 등

import re

def extract_number1(x):
    personcount_num = None  # 초기화

    personcount = re.findall("\d.*명", x)

    try:
        personcount_num = personcount[0].replace("명", "") # 숫자뒤에 '명'을 삭제
        try:
            personcount_num = int(personcount_num.replace(",", "")) # 천단위 표시 삭제
        except:
            personcount_num = int(personcount_num)
    except:
        pass

    return personcount_num if personcount_num is not None else None
```

```python
personcountlist = ['전체 재학생수','정원내 재학생수', '학생정원', '학생모집정지인원', '졸업자수', '취업자수' ,'제외자수',
                   '전체 입학자수', '정원내 입학자수', '정원내 모집인원','재학생수(교육비)','전임교원수(재학생기준)',
                   '재학생수(전임교원)', '교원 법정정원(재학생기준)','재학생수(장학금)']

for i in personcountlist:
  df3[i] = df3[i].apply(extract_number1)
```

```python
import re

def extract_number2(x):
    ratio_num = None  # 초기화

    ratio = re.findall("\d+\.\d+ %", x)  # 소수점 형태의 숫자 추출

    try:
        ratio_num = ratio[0].replace(" %", "")  # % 기호 제거
        ratio_num = float(ratio_num)  # 소수형으로 변환
    except:
        pass

    return ratio_num

```

```python
ratiolist = ['재학생충원율(정원내)','취업률','신입생충원율(정원내)','전임교원 확보율(재학생기준)']

for i in ratiolist:

  df3[i] = df3[i].apply(extract_number2)
```

```python
import re

def extract_and_clean(x):
    cleaned_value = None

    # 정규식으로 숫자와 소수점만 추출
    numbers = re.findall(r"\d+\.*\d*", x)

    if numbers:
        # 리스트의 모든 요소를 결합하여 문자열로 만듦
        combined = ''.join(numbers)

        try:
            cleaned_value = float(combined)
        except:
            pass

    return cleaned_value
```

```python
moneylist = ['총교육비', '학생 1인당 교육비','장학금 총액', '교내 + 사설 및 기타', '학생 1인당 장학금' ]

for i in moneylist:
  df3[i] = df3[i].apply(extract_and_clean)
```
